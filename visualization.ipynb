```python
import pandas as pd
pd.options.plotting.backend = "plotly"
```

```python
def plot_files(file_names):
    data = []
    for file_name in file_names:
        df = pd.read_csv('output/' + file_name, sep='\t').set_index('Size')
        df.name = file_name.split('.')[0]
        df['Insertion time (ns)'] = df['Insertion time (ns)'] / df.index
        df['Query time (ns)'] = df['Query time (ns)'] / df.index
        df['Overall time (ns)'] = df['Insertion time (ns)'] + df['Query time (ns)']
        data.append(df)

    for col_name in data[0].columns:
        insertion_time = pd.DataFrame()
        for df in data:
            if not ('bucket' in col_name and 'Red' in df.name):
                insertion_time[df.name] = df[col_name]

        # Plot insertion time using plotly library
        fig = insertion_time.plot(title=col_name, markers=True, log_x=True)
        fig.show()

    
```

```python
# Get list of files within a directory
import os
file_names = os.listdir('output')
plot_files(file_names)
```
